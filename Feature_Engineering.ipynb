**Valores nulos**

Se identifican los atributos categoricos nulos de train y de test.

```python
def atributos_categoricos_nan(df):
  atributos_nulos = [atributo for atributo in df.columns if df[atributo].dtypes == 'O' and df[atributo].isna().sum()>0]
  return atributos_nulos
```

```python
atributos_categoricos_nan_train = atributos_categoricos_nan(df_train)
atributos_categoricos_nan_train
```

```python
for atributo in atributos_categoricos_nan_train:
    print("{}: {}% valores nulos".format(atributo,np.round(df_train[atributo].isnull().mean()*100,4)))
```

```python
atributos_categoricos_nan_test = atributos_categoricos_nan(df_test)
atributos_categoricos_nan_test
```

Se crea una funci√≥n que reemplaza los valores 'NaN' por 'nulo' para facilitar el manejo de valores faltantes en el an√°lisis posterior de las variables categoricas.





```python
def reemplazar_atributos_categoricos_nan(df, atributos_categoricos_nan):
  dataset = df.copy()
  dataset[atributos_categoricos_nan] = dataset[atributos_categoricos_nan].fillna('Nulo')
  return dataset
```

```python
train_df = reemplazar_atributos_categoricos_nan(df_train, atributos_categoricos_nan_train)
print("Cantidad de valores NaN en train:",train_df[atributos_categoricos_nan_train].isna().sum().sum())
```

```python
test_df = reemplazar_atributos_categoricos_nan(df_test, atributos_categoricos_nan_test)
print("Cantidad de valores NaN en test:",test_df[atributos_categoricos_nan_test].isna().sum().sum())
```

Ahora se identifican los valores numericos nulos.

```python
def encontrar_atributos_numericos_nan(df):
  atributos_numericos_nan = [atributo for atributo in df.columns if df[atributo].dtypes != 'O' and df[atributo].isna().sum()>0]
  return atributos_numericos_nan

```

```python
atributos_num_nan_train = encontrar_atributos_numericos_nan(train_df)
atributos_num_nan_train
```

```python
atributos_num_nan_test = encontrar_atributos_numericos_nan(test_df)
atributos_num_nan_test
```

Una vez identificados los atributos num√©ricos que poseen valores 'NaN', se crea una funci√≥n para rellenarlos. En este caso, se utiliza la mediana como estrategia, debido a la cantidad valores at√≠picos que se observaron en la parte de EDA, y esta se calcula √∫nicamente a partir de los datos de train para evitar la fuga de datos.

Dentro de la funci√≥n, tambi√©n se crea una nueva columna con el nombre original seguido de '_nan', que toma el valor 1 si el registro ten√≠a un valor 'NaN' en la columna original, con el prop√≥sito de conservar la informaci√≥n sobre los valores faltantes.

```python
def llenar_valores_num_nulos(df, atributos_num_nan, medianas):
    for atributo in atributos_num_nan:
        df[atributo + '_nan'] = np.where(df[atributo].isna(), 1, 0)
        df[atributo].fillna(medianas[atributo], inplace=True)
```

```python
atributos_num_nan = list(set(atributos_num_nan_train).union(set(atributos_num_nan_test)))
medianas = {atributo: train_df[atributo].median() for atributo in atributos_num_nan}
medianas
```

**Manejo de valores raros - categ√≥ricos**

```python
atributos_cat=[feature for feature in train_df.columns if train_df[feature].dtype=='O']
atributos_cat
```

Esta funci√≥n asigna el valor 'rare_var' a todas las categor√≠as cuya frecuencia relativa sea menor al umbral definido, que en este caso es del 1%.

```python
def manejar_valores_raros(train_df, test_df, atributos_categoricos, umbral=0.01):
    train_df = train_df.copy()
    test_df = test_df.copy()

    for feature in atributos_categoricos:
        temp = train_df.groupby(feature)['SalePrice'].count() / len(train_df)
        categorias_frecuentes = temp[temp > umbral].index
        train_df[feature] = np.where(train_df[feature].isin(categorias_frecuentes), train_df[feature], 'Rare_var')
        test_df[feature] = np.where(test_df[feature].isin(categorias_frecuentes), test_df[feature], 'Rare_var')

    return train_df, test_df
```

```python
dataset_train, dataset_test = manejar_valores_raros(train_df, test_df, atributos_cat)
```

```python
for atributo in atributos_cat:
  print("Valores √∫nicos para {}: {}".format(atributo, dataset_train[atributo].unique()))
```

```python
full_df = pd.concat([dataset_train, dataset_test])
```

Se define la funci√≥n llamada codificar_orden_categorico que codifica variables categ√≥ricas ordinales bas√°ndose en el promedio de 'SalePrice'. Para cada atributo categ√≥rico, se calcula la media de 'SalePrice' agrupada por las categor√≠as en el DataFrame de train, y estas se ordenan de manera ascendente. Luego, se asigna un valor num√©rico a cada categor√≠a seg√∫n su posici√≥n en el orden. Finalmente, se aplica el mismo mapeo tanto en train como en test para garantizar la consistencia.

```python
def codificar_orden_categorico(train_df, test_df, atributos_categoricos):
    train_df = train_df.copy()
    test_df = test_df.copy()

    for feature in atributos_categoricos:
        labels_ordered = train_df.groupby([feature])['SalePrice'].mean().sort_values().index
        labels_ordered = {k: i for i, k in enumerate(labels_ordered, 0)}
        train_df[feature] = train_df[feature].map(labels_ordered)
        test_df[feature] = test_df[feature].map(labels_ordered)

    return train_df, test_df
```

```python
data_train, data_test = codificar_orden_categorico(dataset_train, dataset_test, atributos_cat)
```

**Estandarizaci√≥n de valores**

Se escalan los atributos para normalizar sus rangos y mejorar el rendimiento del modelo. En este caso, se utiliza MinMaxScaler porque demostr√≥ un mejor ajuste en comparaci√≥n con StandardScaler al mantener los valores dentro de un rango espec√≠fico y evitar la influencia excesiva de los valores extremos.

```python
atributos_a_escalar=[feature for feature in data_train.columns if feature not in ['Id','SalePrice']]
atributos_a_escalar
```

```python
scaler = MinMaxScaler()
scaler.fit(data_train[atributos_a_escalar])
```

```python
train_escalado = pd.DataFrame(scaler.transform(data_train[atributos_a_escalar]), columns=atributos_a_escalar)
```

```python
test_escalado = pd.DataFrame(scaler.transform(data_test[atributos_a_escalar]), columns=atributos_a_escalar)
```

```python
escalado_train = pd.concat([data_train[['Id', 'SalePrice']].reset_index(drop=True), train_escalado], axis=1)
escalado_test = pd.concat([data_test[['Id']].reset_index(drop=True), test_escalado], axis=1)
```

Se llena con -1 el valor que faltaba en la columna 'KitchenQual' porque representa una categor√≠a desconocida, permitiendo al modelo identificar estos casos sin introducir sesgos ni afectar las categor√≠as existentes.

```python
escalado_test['KitchenQual'] = escalado_test['KitchenQual'].fillna(-1)
```
